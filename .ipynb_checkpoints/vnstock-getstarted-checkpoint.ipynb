```python
import sys
print(sys.version)
```

```python
from vnstock import Vnstock
from datetime import datetime
stock = Vnstock().stock(symbol='VCI', source='VCI')
start='2020-01-01'
end=datetime.today().strftime('%Y-%m-%d')
# stock.quote.history(start='2020-01-01', end='2024-05-25')
price_data = stock.quote.history(start=start, end=end)
```

```python
price_data.info()
price_data.columns
price_data.tail()
# price_data.reset_index(drop=True, inplace=True)
# price_data.columns
# price_data.tail()
price_data.set_index(price_data.columns[0], inplace=True)
price_data.index.name = 'Date'
price_data.tail()
```

```python
import bt
print(bt.__version__)
```

```python
# create the strategy
s = bt.Strategy('s1', [bt.algos.RunMonthly(),
                       bt.algos.SelectAll(),
                       bt.algos.WeighEqually(),
                       bt.algos.Rebalance()])
```

```python
# create a backtest and run it
test = bt.Backtest(s, price_data)
```

```python
res = bt.run(test)
```

```python
# first let's see an equity curve
res.plot();
```

```python
# ok and what about some stats?
res.display()
```

```python
# ok and how does the return distribution look like?
res.plot_histogram()
```

```python
# and just to make sure everything went along as planned, let's plot the security weights over time
res.plot_security_weights()
```

```python
# create our new strategy
s2 = bt.Strategy('s2', [bt.algos.RunWeekly(),
                        bt.algos.SelectAll(),
                        bt.algos.WeighInvVol(),
                        bt.algos.Rebalance()])
```

```python
# now let's test it with the same data set. We will also compare it with our first backtest.
test2 = bt.Backtest(s2, price_data)
```

```python
# we include test here to see the results side-by-side
res2 = bt.run(test, test2)
```

```python
res2.plot();
```
